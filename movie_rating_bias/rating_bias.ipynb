# Movie Review Site Bias

Introduce data from FiveThiryEight, which compiled movie data for 147 films in 2015 from movie rating sites: Rotten Tomatoes, Metacritic, IMDB, and Fandango

```python
import pandas as pd
reviews = pd.read_csv("fandango_scores.csv")
list_columns = ["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]
norm_reviews = reviews[list_columns]
norm_reviews[:1]
```

# Bar chat of average user rating from different sources for "Avengers: Age of Ultron (2015)

```python
import matplotlib.pyplot as plt
from numpy import arange
bar_positions = arange(5) 
# Heights of the bars.  In our case, the average rating for the first movie in the dataset.
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values

tick_positions = range(0,5)
fig,ax = plt.subplots()
# width of vars set to 0.5 instead of the default value 0.8
ax.bar(bar_positions,bar_heights,0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols,rotation=90)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

# Horizontal bar chat of average user rating from different sources for "Avengers: Age of Ultron (2015)

```python
fig, ax = plt.subplots()
bar_widths = norm_reviews[num_cols].iloc[0].values
bar_postitions = arange(5)
tick_positions = range(0,5)
ax.barh(bar_postitions,bar_widths,0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_xlabel("Average Rating")
ax.set_ylabel("Rating Source")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

From the bar chats, it is shown that the ratings from Fandango are higher than the ratings from other sources

# Scatter plot to visualize the relationship between the ratings of Fandango and Rotten Tomatoes

```python
fig,ax = plt.subplots()
ax.scatter(norm_reviews["Fandango_Ratingvalue"],norm_reviews["RT_user_norm"])
ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatoes")
plt.show()
```

From the scatter plot, it is shown that there is a weak positive correlation between the ratings of Fandango and that of Rotten Tomatoes.

```python
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.scatter(norm_reviews["Fandango_Ratingvalue"],norm_reviews["RT_user_norm"])
ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatoes")

ax2.scatter(norm_reviews["RT_user_norm"],norm_reviews["Fandango_Ratingvalue"])
ax.set_xlabel("Rotten Tomatoes")
ax.set_ylabel("Fandango")
plt.show()
```

# Scatter plots to investigate the correlation of Fandango ratings with other review sites

```python
fig = plt.figure(figsize=(5,10))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

ax1.scatter(norm_reviews["Fandango_Ratingvalue"],norm_reviews["RT_user_norm"])
ax1.set_xlabel("Fandango")
ax1.set_ylabel("Rotten Tomatoes")
ax1.set_xlim(0,5)
ax1.set_ylim(0,5)

ax2.scatter(norm_reviews["Fandango_Ratingvalue"],norm_reviews["Metacritic_user_nom"])
ax2.set_xlabel("Fandango")
ax2.set_ylabel("Metacritic")
ax2.set_xlim(0,5)
ax2.set_ylim(0,5)

ax3.scatter(norm_reviews["Fandango_Ratingvalue"],norm_reviews["IMDB_norm"])
ax3.set_xlabel("Fandango")
ax3.set_ylabel("IMDB")
ax3.set_xlim(0,5)
ax3.set_ylim(0,5)

plt.show()
```

Ratings from from Fandango and IMDB positively correlate the most. The ratings from Rotten Tomatoes and Metacritic have a weak positive correlation. Furthermore, the range of rating from Rotten Tomatoes and Metacritic varies more than that of IMDB and Fandango

# Histograms and box plots of the movie rating data

```python
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
norm_reviews = reviews[cols]
norm_reviews[:5]
```

# Frequency Distribution

```python
fandango_distribution = norm_reviews["Fandango_Ratingvalue"].value_counts()
imdb_distribution = norm_reviews["IMDB_norm"].value_counts()
fandango_distribution = fandango_distribution.sort_index()
imdb_distribution = imdb_distribution.sort_index()
print(fandango_distribution)
print(imdb_distribution)
```

The Fandango rating ranges from 2.7 to 4.8, while that from IMDB franges from 2 to 4.3. In order to compare the frequency appropriately, the frequency distributions are then sorted into bins.

# Histograms of the movie rating data to visualize the ratings in specific bins

```python
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews["Fandango_Ratingvalue"],20,range=(0,5))
ax1.set_title("Distribution of Fandango Ratings")
ax1.set_ylim(0,50)

ax2.hist(norm_reviews["RT_user_norm"],20,range=(0,5))
ax2.set_title("Distribution of Rotten Tomatoes Ratings")
ax2.set_ylim(0,50)

ax3.hist(norm_reviews["Metacritic_user_nom"],20,range=(0,5))
ax3.set_title("Distribution of Metacritic Ratings")
ax3.set_ylim(0,50)

ax4.hist(norm_reviews["IMDB_norm"],20,range=(0,5))
ax4.set_title("Distribution of IMDB Ratings")
ax4.set_ylim(0,50)

plt.show()
```

# Box plot of the movie rating data to visualize the quartiles of the data

```python
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
fig, ax = plt.subplots()
ax.boxplot(norm_reviews[num_cols].values) #nparray.values() is used to return a multi-dimensional numpy array
ax.set_ylim(0,5)
ax.set_xticklabels(num_cols,rotation = 90) #list notation is used inside ()
plt.show()
```

# Findings
The ratings from Rotten Tomatoes and Metacritic span across a larger range of values than those from IMDB and Fandango. The IMDB and Fandango rations are skewed more positively as well.
