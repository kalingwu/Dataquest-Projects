# US Historic Unemployment Rate

```python
import pandas as pd
unrate = pd.read_csv("unrate.csv")
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

# Unemployment Trends in 1948 and 1949

```python
fig = plt.figure(figsize=(12,7))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12])
ax2.plot(unrate["DATE"][12:24], unrate["UNRATE"][12:24])
ax1.set_title("Monthly Unemployment Trends, 1948")
ax2.set_title("Monthly Unemployment Trends, 1949")
plt.show()
```

# Unemployment Trends from 1948 to 1952 shown in different subplots

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    ax.plot(unrate["DATE"][i*12:(i+1)*12], unrate["UNRATE"][i*12:(i+1)*12])
plt.show()
```

# Unemployment Trends of 1948 to 1949 shown in single plot

```python
unrate["MONTH"]=unrate["DATE"].dt.month
fig = plt.figure(figsize = (6,3))
plt.plot(unrate["MONTH"][0:12],unrate["UNRATE"][0:12],c="red")
plt.plot(unrate["MONTH"][12:24],unrate["UNRATE"][12:24],c="blue")
plt.show()
```

# Unemployment Trends of 1948 to 1949 shown in single plot

```python
fig = plt.figure(figsize = (10,6))
color = ["red","blue","green","orange","black"]
for i in range(5):
    legend_label = str(i+1948)
    plt.plot(unrate["MONTH"][i*12:(i+1)*12],unrate["UNRATE"][i*12:(i+1)*12],c=color[i],label=legend_label)
    plt.legend(loc = "upper left")
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
